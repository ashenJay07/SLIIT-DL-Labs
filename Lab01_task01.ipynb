```python
import numpy as np

# Generate a random 4x4 array from an exponential distribution
scale_parameter = 1.0  # Adjust this parameter to control the distribution's scale
random_array = np.random.exponential(scale=scale_parameter, size=(4, 4))

print(random_array)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a random array of size 100000 from a uniform distribution [0, 1)
random_uniform = np.random.rand(100000)

# Generate histograms for the random array and compare with a standard normal distribution
plt.figure(figsize=(10, 6))

# Plot the histogram of the random uniform distribution
plt.hist(random_uniform, density=True, bins=100, histtype="step", color="blue", label="Uniform")

# Plot the standard normal distribution for comparison
x = np.linspace(-2.5, 2.5, 1000)
y = (1/np.sqrt(2*np.pi)) * np.exp(-x**2 / 2)
plt.plot(x, y, color="red", label="Standard Normal")

plt.legend(loc="upper left")
plt.title("Random and Standard Normal Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range of values for X and Y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate the corresponding Z values
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Plot of Z = X**2 + Y**2')

# Add a color bar
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)

# Show the plot
plt.show()
```
